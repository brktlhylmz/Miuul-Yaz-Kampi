```python
#Görev 17: 
df = sns.load_dataset("tips")
df.head()
```

```python
#Görev 18: 
print(df.groupby("time")["total_bill"].sum())
print(df.groupby("time")["total_bill"].mean())
print(df.groupby("time")["total_bill"].min())
print(df.groupby("time")["total_bill"].max())
```

```python
#Görev 19: 
print(df.groupby(["day","time"])["total_bill"].sum())
print(df.groupby(["day","time"])["total_bill"].min())
print(df.groupby(["day","time"])["total_bill"].max())
print(df.groupby(["day","time"])["total_bill"].mean())
```

```python
#Görev 20: 
print("\nToplamı:")
print(df[df["time"]=="Lunch"][df.sex=="Female"][["total_bill","tip"]].sum())
print("\nMinimumu:")
print(df[df["time"]=="Lunch"][df.sex=="Female"][["total_bill","tip"]].min())
print("\nMaksimumu:")
print(df[df["time"]=="Lunch"][df.sex=="Female"][["total_bill","tip"]].max())
print("\nOrtalaması:")
print(df[df["time"]=="Lunch"][df.sex=="Female"][["total_bill","tip"]].mean())
```

```python
#Görev 21: 
df.loc[(df["size"]>3) & (df["total_bill"]>10)].mean()
```

```python
#Görev 22: 
df["total_bill_tip_sum"] = df.apply(lambda df:df["total_bill"]+df["tip"],axis=1)
df.head()
```

```python
#Görev 23: 
female_mean = df[df.sex=="Female"]["total_bill"].mean()
male_mean = df[df.sex=="Male"]["total_bill"].mean()

def ort (df):
    if df["sex"]== "Female":
        if df["total_bill"]<female_mean:
            return 0
        elif df["total_bill"]>=female_mean:
            return 1
    else:
        if df["total_bill"]<male_mean:
            return 0
        elif df["total_bill"]>=male_mean:
            return 1

df["total_bill_flag"] = df.apply(ort,axis=1)
print(f"Kadınların Ortalaması: {female_mean}\nErkeklerin Ortalaması: {male_mean}")
df
```
